```python
import pandas as pd
new_df = pd.DataFrame(columns=['country', 'year', 'month', 'mean_temp'])
```

```python
file_name = ['brazil', 'colombia', 'congo(democratic)', 'congo(republic of)', 'ecuador',\
            'gabon', 'indonesia', 'kenya', 'kiribati', 'maldives', 'sao tome & principe', 'somalia',\
            'uganda']
rename = {"Jan":'01_Jan', 'Feb':'02_Feb', 'Mar':'03_Mar', 'Apr':'04_Apr',\
         'May':'05_May', 'Jun':'06_Jun', 'Jul':'07_Jul', 'Aug':'08_Aug',\
         'Sep':'09_Sep', 'Oct':'10_Oct', 'Nov':'11_Nov', 'Dec':'12_Dec'}

for i in range(13):
    new_file = pd.read_csv((file_name[i]+'.csv'), encoding='unicode_escape', index_col=0)

    for lab, row in new_file.iterrows():
        for col, _ in new_file.iteritems():
            new_row = {'country':[file_name[i]],\
                       'year':[lab], 'month':[col], 'mean_temp':[row[col]]}
            new_row = pd.DataFrame(new_row)
            new_df = pd.concat([new_df, new_row], ignore_index=True)
for lab, row in new_df.iterrows():
    new_df.loc[lab, 'month'] = rename[row['month']]

new_df
```

```python
new_df.to_csv('mean_temp_country.csv', index=False)
```
