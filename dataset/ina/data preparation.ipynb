```python
import pandas as pd

new_file = pd.read_csv('mean_temp.csv', encoding='unicode_escape', index_col=0)
mean_temp = pd.DataFrame(columns=['year', 'month', 'mean_temp'])
rename = {"Jan":'01_Jan', 'Feb':'02_Feb', 'Mar':'03_Mar', 'Apr':'04_Apr',\
         'May':'05_May', 'Jun':'06_Jun', 'Jul':'07_Jul', 'Aug':'08_Aug',\
         'Sep':'09_Sep', 'Oct':'10_Oct', 'Nov':'11_Nov', 'Dec':'12_Dec'}

for lab, row in new_file.iterrows():
    for col, _ in new_file.iteritems():
        new_row = {'year':[lab], 'month':[col], 'mean_temp':[row[col]]}
        new_row = pd.DataFrame(new_row)
        mean_temp = pd.concat([mean_temp, new_row], ignore_index=True)
mean_temp
```

```python
new_file = pd.read_csv('max_temp.csv', encoding='unicode_escape', index_col=0)
max_temp = pd.DataFrame(columns=['year', 'month', 'max_temp'])

for lab, row in new_file.iterrows():
    for col, _ in new_file.iteritems():
        new_row = {'year':[lab], 'month':[col], 'max_temp':[row[col]]}
        new_row = pd.DataFrame(new_row)
        max_temp = pd.concat([max_temp, new_row], ignore_index=True)
max_temp
```

```python
new_file = pd.read_csv('min_temp.csv', encoding='unicode_escape', index_col=0)
min_temp = pd.DataFrame(columns=['year', 'month', 'min_temp'])

for lab, row in new_file.iterrows():
    for col, _ in new_file.iteritems():
        new_row = {'year':[lab], 'month':[col], 'min_temp':[row[col]]}
        new_row = pd.DataFrame(new_row)
        min_temp = pd.concat([min_temp, new_row], ignore_index=True)
min_temp
```

```python
new_file = pd.read_csv('precipitation.csv', encoding='unicode_escape', index_col=0)
precipitation = pd.DataFrame(columns=['year', 'month', 'precipitation'])

for lab, row in new_file.iterrows():
    for col, _ in new_file.iteritems():
        new_row = {'year':[lab], 'month':[col], 'precipitation':[row[col]]}
        new_row = pd.DataFrame(new_row)
        precipitation = pd.concat([precipitation, new_row], ignore_index=True)
precipitation
```

```python
df_ina = mean_temp.copy()
df_ina['max_temp'] = max_temp['max_temp']
df_ina['min_temp'] = min_temp['min_temp']
df_ina['precipitation'] = precipitation['precipitation']
df_ina
```

```python
df_ina.to_csv('df_ina.csv', index=False)
```
